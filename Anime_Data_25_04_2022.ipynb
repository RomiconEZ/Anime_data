```python
def inspect_data(dataset):
    problem_cells = sum(dict(dataset.isna().sum()).values()) # the total number of cells containing NaNs
    if problem_cells > 0:
        print("contains some NaNs:")
        print({key:value for key,value in dataset.isna().sum().items() if value>0})
    else:
        print("does not contain any NaNs")  
```

```python
inspect_data(dataset)
```

```python
dataset['Episodes'] = dataset['Episodes'].apply(pd.to_numeric, errors='coerce')
dataset['Voters'] = dataset['Voters'].apply(pd.to_numeric, errors='coerce')
```

```python
cat_cols = [
    'Title',
    'Production',
    'Source',
    'Genre',
    'Airdate',
    'Theme'
]
```

```python
"""Посчитаем значения категориальных признаков"""
for feature in cat_cols:
    print(f'Значения категориального признака {feature}:')
    print(dataset[feature].value_counts())
    print('*************************') 
```

```python
for i, row in dataset.iterrows():
    if dataset.iloc[i].isna().sum()>0:
        print(f"Index: {i}")
        print(f"{row}\n")
```

```python
from sklearn.impute import SimpleImputer
```

```python
# для удобства работы с нашим датасетом создаем маску, указывающую на столбцы с категориальными признаками
cat_features_mask = (dataset.dtypes == "object").values 
# для вещественнозначных признаков заполним пропуски средними
dataset_real = dataset[dataset.columns[~cat_features_mask]]
mis_replacer = SimpleImputer(missing_values=np.nan,strategy='mean')
dataset_no_mis_real = pd.DataFrame(data=mis_replacer.fit_transform(dataset_real), columns=dataset_real.columns)
# для категориальных - пустыми строками
dataset_cat = dataset[dataset.columns[cat_features_mask]].fillna("")

dataset_no_mis = pd.concat([dataset_no_mis_real, dataset_cat], axis=1)
```

```python
dataset_no_mis.info()
```

```python
dataset_no_mis.columns=dataset_no_mis.columns.str.replace(' ','_')
```

```python
dataset_no_mis.columns=dataset_no_mis.columns.str.lower()
```
